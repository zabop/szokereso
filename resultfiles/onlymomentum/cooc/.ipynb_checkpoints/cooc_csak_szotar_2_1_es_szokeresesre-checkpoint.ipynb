```python
import pandas as pd
import glob
from tqdm import tqdm
tqdm.pandas()
```

```python
resfiles = glob.glob('/mnt/volume/jupyter/szokereso/resultfiles/onlymomentum/cooc/data_*_szokereso_result_szotar2_1_only.csv')
```

```python
dictstrings = ['szotar_2.1']
```

```python
len(resfiles)
```

```python
res={}
for index, resfile in enumerate(resfiles):
    if index%30==0: print(index)
    df=pd.read_csv(resfile)
    freqdf = pd.concat([df[col]
                       for dictstring in dictstrings
                       for col in df.columns if dictstring in col]
                       ).value_counts(
                       ).rename_axis('unique_entries'
                       ).reset_index(name='counts')
    res[resfile]=freqdf
```

```python
concated_df=\
pd.concat([res[file] for file in res.keys()]
         ).groupby('unique_entries'
         ).agg({'counts': 'sum'}
         ).sort_values(by=['counts'],ascending=False
         ).reset_index()
```

```python
cooc = pd.DataFrame(index=concated_df['unique_entries'], columns=concated_df['unique_entries'])
cooc.index.names =   ['row']
cooc.columns.names = ['column']
```

```python
dictcols = [col for col in df.columns if any([dictstring in col for dictstring in dictstrings])]
```

```python
len(cooc.index)
```

```python
for icoocrow, coocrow in enumerate(cooc.index):
    #if icoocrow % 30 == 0: print(icoocrow)
    for icooccol, cooccol in enumerate(cooc.columns):
        if icoocrow ==icooccol: print(icoocrow/len(cooc.index))
        if icoocrow > icooccol:
            count=0
            for index, row in df[[*dictcols]].iterrows():
                if cooccol in list(row) and coocrow in list(row):
                    count+=1
            cooc.loc[coocrow,cooccol]=count
```

```python
#!!!RUN ONLY ONCE!!!
cooc=cooc.add(cooc.transpose(),fill_value=0)
```

```python
cooc.to_csv('cooc_csak_szotar_2_1_es_szokeresesre.csv')
```

```python
f = open("cooc_szotar_2_1_only.txt", "w")
for i, col in enumerate(cooc.columns):
    for j, index in enumerate(cooc.index):
        if i!=j:
            sig=cooc.loc[col,index]
            if sig != 0:
                print(col, index, sig)
                f.write('|'.join([str(each) for each in [col, index, sig, '\n']]))
f.close()
```
