```python
df = persondatadict.name_list.str.split('|',expand=True).rename(columns=(lambda num: 'alias'+repr(num)))
```

```python
drFiltered=\
df[[col for col in df.columns if 'alias' in col] # take each column with the string "alias" in them
    ].apply(lambda r: pd.Series( # take each row as input of a function, the result will be put in a series
            name.replace('Dr. ','').replace('dr. ','') for name in r # iterate through contents of the row, replace "Dr. " & "dr. " with nothin
            if type(name) is not float and name is not None),axis='columns') # only do this when the cell content is not None or float, apply apply() along each column ie to each row
```

```python
drFiltered = drFiltered.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
# ^^^^ to get rid of leading/trailing spaces in dicts like ' Nagy Mária' or 'Nagy Mária ', using https://stackoverflow.com/a/45355563/8565438
```

```python
drFilteredUniquized = pd.DataFrame(list(map(pd.unique, drFiltered.values))
# ^^^^ each row should contain one value only once, using: https://stackoverflow.com/a/63583984/8565438
                                   ).rename(columns=(lambda col: 'alias'+repr(col)))
                                   # ^^^^ rename columns
```

```python
newPersonDataDict = \
personDataDict[['personid']
              ].join(drFilteredUniquized
              ).join(persondatadict[['szulido','birthlocation','halalido','deathlocation','profession_list']])
```
