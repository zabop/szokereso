```python
import re
import time
import matplotlib.pyplot as plt

def find(string, text):
    if string.find(text) > -1:
        pass

def re_find(string, text):
    if re.match(text, string):
        pass

def in_find(string, text):
    if text in string:
        pass
```

```python
options = [find, re_find, in_find]
```

```python
import string
import random
random.seed=42
text=' '.join([''.join(random.sample(string.ascii_letters,random.randrange(1,20))) for _ in range(100)])
```

```python
elapsed={function.__name__:[] for function in options}
for string in text.split(' '):
    for function in options:
        start=time.time()
        function(text,string)
        end = time.time()
        elapsed[function.__name__].append(end-start)
```

```python
plt.yscale('log')
for method in elapsed:
    plt.plot(elapsed[method],label=method)
plt.legend()
```
